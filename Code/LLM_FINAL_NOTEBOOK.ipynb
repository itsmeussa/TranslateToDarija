Here we gonna structure and clean our Json files for avoiding problems during finetunning

```python
def is_valid_translation(translation):
    # Check if the translation object contains all required keys
    required_keys = ['idx', 'English', 'Darija', 'MSA']
    return all(key in translation for key in required_keys)

def clean_json(input_file, output_file):
    with open(input_file, 'r') as f_in, open(output_file, 'w') as f_out:
        for line in f_in:
            try:
                translation = json.loads(line)
                if is_valid_translation(translation['translation']):
                    f_out.write(line)
            except json.JSONDecodeError:
                # If a line is not valid JSON, skip it
                pass

#For first dataset
input_file = 'train-16k.json'
output_file = 'train-gpt-16k.json'
clean_json(input_file, output_file)
print("Cleaning finished.")
#For second dataset
input_file = 'online-train-16k.json'
output_file = 'train-huggingface-16k.json'
clean_json(input_file, output_file)
print("Cleaning finished.")
```

```python
# Define the paths to the source files and the destination file
source_files = [
    '/content/drive/MyDrive/LLM Project/train-6000.json',
    '/content/drive/MyDrive/LLM Project/train-online-10k.json',
    '/content/drive/MyDrive/LLM Project/train-10k.json'
]
destination_file_path = '/content/drive/MyDrive/LLM Project/train_data-15k.json'

# Initialize an empty list to store lines
merged_lines = []

# Iterate over each source file to read the first 5000 lines
for file_path in source_files:
    with open(file_path, 'r') as file:
        for _ in range(5000):  # Read first 5000 lines
            line = file.readline()
            # Break if the file has fewer than 5000 lines
            if not line:
                break
            merged_lines.append(line)

# Write the merged lines to the destination file
with open(destination_file_path, 'w') as file:
    file.writelines(merged_lines)

print("Files have been successfully merged.")
```

```python
from collections import deque
import itertools

# Define the paths to the source files and the destination file
source_files = [
    '/content/drive/MyDrive/LLM Project/train-6000.json',
    '/content/drive/MyDrive/LLM Project/train-online-10k.json',
    '/content/drive/MyDrive/LLM Project/train-10k.json'
]
destination_file_path = '/content/drive/MyDrive/LLM Project/test_data_3k.json'

# Function to read the last N lines from a file
def read_last_n_lines(file_path, n):
    with open(file_path, 'r') as file:
        # Using deque to efficiently append to the left and maintain a maximum length
        return deque(file, maxlen=n)

# Use the function to read the last 1000 lines from each file
last_lines_from_each_file = itertools.chain.from_iterable(read_last_n_lines(path, 1000) for path in source_files)

# Write the merged lines to the destination file
with open(destination_file_path, 'w') as file:
    file.writelines(last_lines_from_each_file)

print("Successfully merged the last 1000 lines from each file.")
```
